```python
import matplotlib.pyplot as plt, numpy as np
```

```python
l1_lst_means = []; l2_lst_means = []; l1_lst_sd = []; l2_lst_sd = []
for i in range(11):
  l1 = []; l2 = []
  r = np.random.rand(100, 2**i)
  for j in range(100):
    for k in range(j+1,100):
      l1.append(np.linalg.norm(r[j]-r[k], 1)); l2.append(np.linalg.norm(r[j]-r[k])**2)
  l1_lst_means.append(np.mean(l1)); l2_lst_means.append(np.mean(l2))
  l1_lst_sd.append(np.std(l1)); l2_lst_sd.append(np.std(l2))
```

```python
size = [1,2,4,8,16,32,64,128,256,512,1024]
plt.plot(size,l1_lst_means,label='L1 distance'); plt.plot(size,l2_lst_means,label='Squared L2 distance')
plt.legend()
plt.xlabel("Dimension"); plt.ylabel("Mean of distance between all points"); plt.title("Mean of distance of 100 sampled points vs. dimension")
plt.show()
```

```python
plt.plot(size,l1_lst_sd,label='L1 distance'); plt.plot(size,l2_lst_sd,label='Squared L2 distance')
plt.legend()
plt.xlabel("Dimension"); plt.ylabel("Standard deviation of distance between all points"); plt.title("Standard deviation of distance of 100 sampled points vs. dimension")
plt.show()
```
